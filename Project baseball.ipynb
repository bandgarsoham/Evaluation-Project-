### Removing OUtlier

```python
#Removing OUtlier
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
df_new=df[(z<3).all(axis=1)]
```

### Log Trasformation method for Skewness Removing

```python
from scipy.stats import boxcox # Remove Skewness from some columns using o= Log Trasformation method
df_new['H']=boxcox(df_new['H'],0)
df_new['SO']=boxcox(df_new['SO'],0)
df_new['RA']=boxcox(df_new['RA'],0) 
df_new['ER']=boxcox(df_new['ER'],0) 
df_new['BB']=boxcox(df_new['BB'],0) 
df_new['R']=boxcox(df_new['R'],0) 

#df_new.insert(len(df.columns),'A_Sqrt',np.sqrt(df_new.iloc[:,0]))
```

### Train Test Split
     separating the columns into features and target:

```python
x=df_new.drop(columns=['W'],axis=1)
y=df_new.iloc[:,0]
```

### Scaling the data using Min-Max Scaler:

```python
from sklearn.preprocessing import MinMaxScaler #to equilise all dataset for higher accuracy
mns=MinMaxScaler()
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Find the  beast Random_state use here (currently we are moving shead with Random_state 12)

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=12)
```

```python
print(x.shape,x_train.shape,x_test.shape)
print(y.shape,y_train.shape,y_test.shape)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train) #we got Model learning score is very best which is best for further Prediction
```

```python
predlr=lr.predict(x_test)
```

```python
print(r2_score(y_test,predlr))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,predlr))
print('Mean squared error:',mean_squared_error(y_test,predlr))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predlr)))
```

### Regularization: 
### GridSearchCV
### Cross Validation Score

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression,Ridge,Lasso

import warnings
warnings.filterwarnings('ignore')
```

```python
parameters={'alpha':[.0001,.001,.01,.1,1,10],'random_state':list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
ls=Lasso(alpha=0.01,random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
pred_ls=ls.predict(x_test)

lss=r2_score(y_test,pred_ls)
lss
```

```python
cv_score=cross_val_score(ls,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean
```

```python
parameters={'alpha':[.0001,.001,.01,.1,1,10],'random_state':list(range(0,10))} #For Ridge 
rd=Ridge()
clf=GridSearchCV(rd,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
rd=Ridge(alpha=0.01,random_state=0)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
pred_rd=rd.predict(x_test)

rds=r2_score(y_test,pred_rd)
rds
```

```python
cv_score=cross_val_score(rd,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean
```

### KNeighborsRegressor : Model,GridSearchCV,Cross Validation Score

```python
from sklearn.neighbors import KNeighborsRegressor
parameters={'n_neighbors':[1,2,3,4,5],'algorithm':['auto','ball_tree','kd_tree','brute']}
knn=KNeighborsRegressor()
grid=GridSearchCV(knn,parameters)  
grid.fit(x_train,y_train) 
#print(grid.best_Score)           
#print(grid.best_estimator.n_neighbors)
print(grid.best_params_)            
```

```python
knn=KNeighborsRegressor(n_neighbors=5,algorithm='auto')
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
pred_knn=knn.predict(x_test)

knns=r2_score(y_test,pred_knn)
print('R2 Score:',knns*100)

knnscore=cross_val_score(knn,x,y,cv=5)
knnc=knnscore.mean()
print('Cross Val Score :',knnc)
```

### Support Vector Regressor : Model,GridSearchCV,Cross Validation Score

```python
from sklearn.svm import SVR

parameters={'kernel':['linear','poly','rbf','sigmoid']}
svr=SVR()
grid=GridSearchCV(svr,parameters)  
grid.fit(x_train,y_train) 
#print(grid.best_Score)           
#print(grid.best_estimator.n_neighbors)
print(grid.best_params_)            
```

```python
svr=SVR(kernel='linear')
svr.fit(x_train,y_train)
svr.score(x_train,y_train)
pred_svr=svr.predict(x_test)

knns=r2_score(y_test,pred_svr)
print('R2 Score:',knns*100)

svrscore=cross_val_score(svr,x,y,cv=5)
svrc=svrscore.mean()
print('Cross Val Score :',svrc)
```
